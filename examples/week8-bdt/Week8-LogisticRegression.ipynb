```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt

import seaborn as sns
sns.set_theme()

def logist(x, m, c):
  return 1 / (1 + np.exp(-(c+m*x)))

x_data = np.array([0, 3, 6, 7, 9, 10, 12, 13])
y_data = np.array([0, 0, 1, 0, 1, 1, 1, 1])

popt, pcov = opt.curve_fit(logist, x_data, y_data, method="trf")

x_vals = np.linspace(0, 15, 200)
y_vals = logist(x_vals, *popt)

plt.plot(x_data, y_data, 'o', label='data')
plt.plot(x_vals, y_vals, label='Logistic')
plt.legend()
plt.xlabel('x')
plt.ylabel('P(x)')
plt.savefig('logist-example.pdf')
plt.show()




```

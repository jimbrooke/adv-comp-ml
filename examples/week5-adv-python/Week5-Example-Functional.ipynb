## Lambda Functions

Functional programming is supported in Python using the lambda keyword, which provides a programmatic way to create short functions.  

Below is a standard function definition

```python
def function(x):
    return 0.5*x**2 + x + 1
function(4)
```

The equivalent using lambdas is below.  This just creates an "anonymous" function (so-called because it has no name) and immediately calls it with argument x=4.

```python
(lambda x: 0.5*x**2 + x + 1)(4)
```

## Function Composition



```python
import numpy as np
from functools import reduce

x = np.arange(0,1000)

squares = list(map((lambda x: x**2), x))
sum = reduce((lambda x, y: x+y), squares)

print(sum)
```

```python
print(reduce((lambda x, y: x+y), list(map((lambda x: x**2), x))))
```

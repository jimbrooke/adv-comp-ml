# Exceptions

This is a short notebook on exceptions, which are a standard technique for managing errors, using examples from the course notes and lectures.

## Simple error handling

You might have managed errors previously like this :

```python
import math
import numpy as np
```

```python
def mySqrt(x):
    if x<0:
        print("Input must be positive.")
        return np.nan
    return math.sqrt(x)

print(mySqrt(4))
print(mySqrt(-3))
```

Effectively we are alerting the caller of `mySqrt()` that there was an error via the returned value.  This can work, but it relies on the calling code to identify and react to the `np.nan` return value.  Frequently what really happens is the calling code just proceeds using the special return value as a real number, until it crashes.

## Exceptions

Exceptions are basically a standardised format for errors, with some special features which make them more useful than managing errors via the function return value.  You will have seen them in action already, for example :

```python
myFunction()
```

Here, the Python interpreter attempts to run the cell contents, but it cannot find a `myFunction` function, so it raises a `NameError` exception, resulting in the printed error.

We can create an exception in the `mySqrt()` function like this :

```python
def mySqrt(x):
    if x<0:
        raise ValueError("Negative input")
    return math.sqrt(x)

print(mySqrt(4))
print(mySqrt(-4))
```

This uses the builtin exception type `ValueError`, which is _"Raised when an operation or function receives an argument that has the right type but an inappropriate value"._  See https://docs.python.org/3/library/exceptions.html#bltin-exceptions

Note that in this case, the function will abort when the exception is raised, and return to the calling code.  If the calling code cannot handle (or "catch") the exception, then it wil be passed to the function which called that...  until it reachs the top of the "call stack".  You might have seen error output like the below.

```python

def function1():
    function2()
    return

def function2():
    function3()
    return

def function3():
    function4()
    return

def function4():
    function5()
    return

def function5():
    function6()
    return
```

```python
function1()
```

Here there are several nested functions, but `function5()` attempts to call a non-existent function, `function6()`. THis creates an exception which is passed up the chain to the first function called, `function1()` and since none of the functions "catch" the exception, the error is printed.

If `function3()` (say) was able to catch this exception, it might do something different.

```python
def function3():
    try:
        function4()
    except NameError:
        print("Oooops!")
    return
```

```python
function1()
```

This is a trivial/ridiculous example, but illustrates how the code which detects and creates the error can be in quite a different place from the code which knows how to handle the error.  This situation arises frequently in practise, and is a key reason why Exceptions exist.

In the example below, we catch an error of the type above from `mySqrt()` in a function which knows how to manage the square root of negative numbers.

```python
def mySqrtComplex(x):
    try:
        y = mySqrt(x)
    except ValueError:
        y = 1j * mySqrt(abs(x))
    return y

print(mySqrtComplex(16))
print(mySqrtComplex(-16))
```

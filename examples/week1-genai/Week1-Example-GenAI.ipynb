### The Problem - Damped Oscillator

The problem we will consider is solving a damped harmonic oscillator, described by the equation below :

$$
m \frac{d^2x}{d t^2} + c\frac{dx}{dt} + kx = 0
$$

with constants :
   * Mass $m = 1$
   * Damping constant $c = 0.2$
   * Spring constant $k = 2.0$

And initial conditions :
   * $x(0) = 1$
   * $\frac{dx}{dt}(0) = 0$
